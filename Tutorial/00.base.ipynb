## 1.读取并显示图片

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("imori.jpg")
# cv2.imshow("sample", img)  # jupyter不能直接用cv2来imshow，会crash
plt.imshow(img)
plt.show()
```

## 2. 图片基本信息

```python
# Height, Wide, Channel = img.shape
# 所以坐标系为：Height=y, Wide=x, Channel=z
# 通道顺序为BGR而不是RGB 
H, W, C = img.shape
print(H, W, C)
```

```python
print(type(img), img.dtype)
```

# 练习问题
将图像的左半部分的红通道和蓝通道交换

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("imori.jpg")
W_half = img.shape[1] // 2
# BGR: B-->R, R-->B
img[:, :W_half] = img[:, :W_half, (2, 1, 0)]

plt.imshow(img)
plt.show()
```
